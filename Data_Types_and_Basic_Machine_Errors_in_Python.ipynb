An example of when you might encounter truncation errors is when solving integrals (a sequence of summations) on a computer. Given that we mostly work with computers that have finite precision arithematic, we must represent integrals as finite summations.

To demonstrate truncation errors we are going to sum all the integer values from $1$ to $N$, but were are going to represent each integer as a numpy.float32 type. In the cell below, I've writen a function that finds the analyical solution to $\sum\limits_{i=1}^{N}i$, for a given $N$.

* Once you have verified that your sumation function is returning the correct results, you have two tasks:

````python
import numpy as np

# a function that returns the analytical sum of the numbers 1 through N
def analytic_sum(N):
    return np.array([0.5*N*(N+1)]).astype(np.dtype(float))[0]


# your summation function that takes numpy array of type numpy.float32
def my_sum(myarr):
    return np.sum(myarr)


# your improved summation function 
def my_impr_sum(myarr):
    
    x = myarr.astype(np.int32)
    return np.sum(x)
 

# Special Case: recursive summation function here
def recur_sum(myarr, n):    
    if len(myarr) == 1:
        return myarr[0] // 1
    
    return myarr[n-1]//1 + recur_sum(myarr[:n-1], n-1)



# put your code to test the functions here. Use comments!

# create a Numpy array of size N
N = 20001
x = np.arange(1, N+1, 1, dtype = np.float32)

# result of my summation function
mysum = my_sum(x)
ansum = analytic_sum(N)
print(mysum)
print(ansum)
print('equals to analytical result? ', mysum==ansum)
print()

# result of analytical summation function
analsum = analytic_sum(N)

# result of improved summation function 
N = 20001 * 2
x = np.arange(1, N+1, 1, dtype = np.float32)

imsum = my_impr_sum(x)
ansum = analytic_sum(N)
print(imsum)
print(ansum)
print('equals to analytical result? ', imsum==ansum)
print()

# result of recursive sum function
N = 20001
x = np.arange(1, N+1, 1, dtype = np.float32)
ansum = analytic_sum(N)

import sys
sys.setrecursionlimit(N + 1000)
recursum = recur_sum(x, N)

print(recursum)
print(ansum)
print('equals to analytical result? ', recursum == ansum)
````
